p.s. –∑–¥–æ—Ä–æ–≤–æ, —á—Ç–æ –µ—Å—Ç—å –æ–±—ä—è—Å–Ω–µ–Ω–∏–µ –ø–æ–≤–µ–¥–µ–Ω–∏—è –¥–∞–Ω–Ω–æ–≥–æ –≥—Ä–∞—Ñ–∏–∫–∞. –ü—ã—Ç–∞—Ç—å—Å—è –ø–æ–Ω—è—Ç—å –ø–æ–ª—É—á–µ–Ω–Ω—ã–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç - –≤–∞–∂–Ω–∞—è —á–∞—Å—Ç—å —Ä–∞–±–æ—Ç—ã –∞–Ω–∞–ª–∏—Ç–∏–∫–∞.</div>


```python
df = df[df["year_of_release"] >= 1995]
# –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã —Å –Ω–∞–∏–±–æ–ª—å—à–∏–º–∏ —Å—É–º–º–∞—Ä–Ω—ã–º–∏ –ø—Ä–æ–¥–∞–∂–∞–º–∏
platform_total_sales = (
    df.pivot_table(index="platform", values="total_sales", aggfunc="sum")
    .sort_values(by="total_sales", ascending=False)
)
# platform = platform[platform["total_sales"] > 0]
platform_total_sales.head(20)
```

```python
# —Ç–∞–±–ª–∏—Ü–∞ —Å —Å—É–º–º–∞—Ä–Ω—ã–º–∏ –ø—Ä–æ–¥–∞–∂–∞–º–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º –ø–æ –≥–æ–¥–∞–º
platform_sales = df.pivot_table(index="platform",columns="year_of_release", values="total_sales", aggfunc="sum")
# –æ–±—ä–µ–¥–∏–Ω—è–µ–º —Å —Ç–∞–±–ª–∏—Ü–µ–π-—Ä–µ–π—Ç–∏–Ω–≥–æ–º –ø—Ä–æ–¥–∞–∂
platform_pivot = platform_total_sales.join(platform_sales).drop("total_sales", axis=1)
# platform = platform.fillna(0)
# —Å –Ω–µ–∑–∞–ø–æ–ª–Ω–µ–Ω–Ω—ã–º–∏ –ø—Ä–æ–ø—É—Å–∫–∞–º–∏ —Ö–∏—Ç–º—ç–ø –≤—ã–≥–ª—è–¥–∏—Ç –∫—Ä—É—á–µ
plt.figure(figsize=(20, 10))
sns.heatmap(platform_pivot, annot=True, fmt=".3g", cmap="Blues", linewidths=1, linecolor="gray")
```
