6. —Ä–∞—Å—Å–º–æ—Ç—Ä–∏–º –∫–æ–ª–æ–Ω–∫—É —Å –Ω–∞—Å–µ–ª–µ–Ω–Ω—ã–º –ø—É–Ω–∫—Ç–æ–º. –ï—Å—Ç—å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –Ω–∞–π—Ç–∏ –≤ –¥–∞—Ç–∞—Å–µ—Ç–µ –ø–æ–¥—Å–∫–∞–∑–∫–∏ –¥–ª—è –∑–∞–ø–æ–ª–Ω–µ–Ω–∏—è –ø—Ä–æ–ø—É—Å–∫–æ–≤ –≤ —ç—Ç–æ–π –∫–æ–ª–æ–Ω–∫–µ.

—ç—Ç–æ –≤—Å–µ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ –æ–¥–Ω–æ –∏ —Ç–æ –∂–µ –º–µ—Å—Ç–æ. –≤—Å–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è —Å–æ–≤–ø–∞–¥–∞—é—Ç, –∑–Ω–∞—á–∏—Ç –Ω–∞—Å–µ–ª–µ–Ω–Ω—ã–π –ø—É–Ω–∫—Ç - –ü–∏—Ç–µ—Ä. –ø–æ–ø—Ä–æ–±—É–µ–º –∞–≤—Ç–æ–º–∞—Ç–∏–∑–∏—Ä–æ–≤–∞—Ç—å –ø–æ–∏—Å–∫ –ø–æ–¥–æ–±–Ω—ã—Ö —Å–ª—É—á–∞–µ–≤

```python
# —Å–æ–∑–¥–∞–¥–∏–º –≤—Å–ø–æ–º–æ–≥–∞—Ç–µ–ª—å–Ω—É—é —Ç–∞–±–ª–∏—Ü—É –∏–∑ —Å—Ç—Ä–æ–∫ —Å –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏–µ–º –Ω–∞—Å–µ–ª–µ–Ω–Ω–æ–≥–æ –ø—É–Ω–∫—Ç–∞
df_locless = df[df["locality_name"].isna()]

def airport(column):
    ''' —Ñ—É–Ω–∫—Ü–∏—è –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Å–ø–∏—Å–æ–∫ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö 
          –∑–Ω–∞—á–µ–Ω–∏–π —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è –¥–æ –∞—ç—Ä–æ–ø–æ—Ä—Ç–∞'''
    air = []
    for item in column:
        if item not in air and item > 0:
            air.append(item)
    return air

air = airport(df_locless["airports_nearest"])   

print(air)
```

```python
# –Ω–∞–π–¥–µ–º —Å—Ä–µ–∑ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞, –≤ –∫–æ—Ç–æ—Ä–æ–º —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –¥–æ –∞—ç—Ä–æ–ø–æ—Ä—Ç–∞ —Å–æ–≤–ø–∞–¥–∞–µ—Ç —Å–æ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –∏–∑ —Å–ø–∏—Å–∫–∞
temp = df.query('airports_nearest in @air')
len(temp)
```

—Å–æ–≤–ø–∞–¥–µ–Ω–∏–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–π –¥–æ  —Ç—Ä–µ—Ö —Ç–æ—á–µ–∫ –≥–∞—Ä–∞–Ω—Ç–∏—Ä—É–µ—Ç —Å–æ–≤–ø–∞–¥–µ–Ω–∏–µ –∞–¥—Ä–µ—Å–∞ (–≤ –º–∞—Å—à—Ç–∞–±–∞—Ö –Ω–∞—Å–µ–ª–µ–Ω–Ω–æ–≥–æ –ø—É–Ω–∫—Ç–∞ —Ç–æ—á–Ω–æ)

```python
# –æ—Ç—Ñ–∏–ª—å—Ç—Ä—É–µ–º —Å—Ç—Ä–æ–∫–∏ —Å—Ä–µ–∑–∞ –≤ –∫–æ—Ç–æ—Ä—ã—Ö –ø—Ä–∏—Å—É—Ç—Å—Ç–≤—É–µ—Ç –∑–Ω–∞—á–µ–Ω–∏–µ –Ω–∞—Å–µ–ª–µ–Ω–Ω–æ–≥–æ –ø—É–Ω–∫—Ç–∞ –∏ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è –¥–æ –¥–≤—É—Ö —Ç–æ—á–µ–∫
temp = temp[(temp["locality_name"].notna()) & (temp["city_centers_nearest"].notna()) & (temp["ponds_nearest"].notna())]

# –ø–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã
print(len(temp["city_centers_nearest"].unique()))
print(len(temp))
temp["locality_name"].unique()
```

 —É –Ω–∞—Å –µ—Å—Ç—å 14 –∞–¥—Ä–µ—Å–æ–≤, 53 –æ–±—ä–µ–∫—Ç–∞ –∏ –≤—Å–µ –≤ –ü–∏—Ç–µ—Ä–µ. —É–¥–æ–±–Ω–æ.

```python
# –¥–æ–±–∞–≤–∏–º –¥–æ—Å—Ç–æ–≤–µ—Ä–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ —Å—Ç–æ–ª–±–µ—Ü —Å –Ω–∞—Å–µ–ª–µ–Ω–Ω—ã–º–∏ –ø—É–Ω–∫—Ç–∞–º–∏
center = list(temp["city_centers_nearest"])
ponds = list(temp["ponds_nearest"])
airports = list(temp["airports_nearest"])
for i in range(len(airports)):
    df.loc[(df["city_centers_nearest"] == center[i]) & (df["ponds_nearest"] == ponds[i]) & (df["airports_nearest"] == airports[i]), "locality_name"] = "–°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥"
# –ø—Ä–æ–≤–µ—Ä—è–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç
df["locality_name"].isna().sum()
```

—Ç–µ–ø–µ—Ä—å –∑–∞–ø–æ–ª–Ω–∏–º –æ—Å—Ç–∞–≤—à–∏–µ—Å—è –ø—Ä–æ–ø—É—Å–∫–∏ –Ω–∞ –ø—É—Å—Ç—É—é —Å—Ç—Ä–æ–∫—É –∏ –ø–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ —É–Ω–∏–∫–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è, –≤–æ–∑–º–æ–∂–Ω–æ —Ç–∞–º –µ—Å—Ç—å –Ω–µ—è–≤–Ω—ã–µ –¥—É–±–ª–∏–∫–∞—Ç—ã

—Å—Ä–µ–¥–∏ –Ω–∞–∏–±–æ–ª–µ–µ —á–∞—Å—Ç–æ –≤—Å—Ç—Ä–µ—á–∞—é—â–∏—Ö—Å—è –µ—Å—Ç—å "–ø–æ—Å—ë–ª–æ–∫ –ú—É—Ä–∏–Ω–æ", —Å—Ä–µ–¥–∏ –Ω–∞—Å–µ–ª–µ–Ω–Ω—ã—Ö –ø—É–Ω–∫—Ç–æ–≤ –±–µ–∑ —É–∫–∞–∑–∞–Ω–∏—è —Ç–∏–ø–∞ –æ–Ω —Ç–∞–∫ –∂–µ –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è.
–ù–µ—è–≤–Ω—ã–µ –¥—É–±–ª–∏–∫–∞—Ç—ã –≤ —ç—Ç–æ–º —Å—Ç–æ–ª–±—Ü–µ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ –µ—Å—Ç—å, —á—Ç–æ –≥–æ–≤–æ—Ä–∏—Ç –æ —Ç–æ–º, —á—Ç–æ —ç—Ç–∏ –¥–∞–Ω–Ω—ã–µ –∑–∞–ø–æ–ª–Ω—è–ª–∏—Å—å –≤—Ä—É—á–Ω—É—é, –Ω–æ –∑–Ω–∞—á–µ–Ω–∏–π –º–Ω–æ–≥–æ –∏ –¥–ª—è —Ç–æ–≥–æ, —á—Ç–æ–±—ã –Ω–∞–π—Ç–∏ –ø—Ä–æ—á–∏–µ –¥—É–±–ª–∏–∫–∞—Ç—ã, —É–¥–æ–±–Ω–µ–µ –≤—Å–µ–≥–æ –±—É–¥–µ—Ç —É–¥–∞–ª–∏—Ç—å –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ —Ç–∏–ø—ã –Ω–∞—Å–µ–ª–µ–Ω–Ω—ã—Ö –ø—É–Ω–∫—Ç–æ–≤, –≥–¥–µ –æ–Ω–∏ —É–∫–∞–∑–∞–Ω—ã 

```python
def separate_locality_name(column):
    '''—Ñ—É–Ω–∫—Ü–∏—è –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Å–ø–∏—Å–æ–∫ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –Ω–∞—Å–µ–ª–µ–Ω–Ω—ã—Ö –ø—É–Ω–∫—Ç–æ–≤,
       —É–¥–∞–ª—è—è –≤—Å–µ —Å–∏–º–≤–æ–ª—ã –¥–æ –ø–µ—Ä–≤–æ–π –∑–∞–≥–ª–∞–≤–Ω–æ–π –±—É–∫–≤—ã –≤ —Å—Ç—Ä–æ–∫–µ'''
    point = []
    word = ""
    for item in column: 
        if item == "":
            continue
        for letter in range(len(item)):
            if item[letter] > "–Ø" or item[letter] == " ":
                continue

            word = item[letter:]
            if word not in point :
                point.append(word)
                word = ""
            break
            
    return point
    
point = separate_locality_name(df["locality_name"])
print(point)
```

```python
def fix_locality_name(locality_name):
    """—Ñ—É–Ω–∫—Ü–∏—è –∑–∞–º–µ–Ω—è–µ—Ç –∑–Ω–∞—á–µ–Ω–∏—è —Å—Ç–æ–ª–±—Ü–∞ –Ω–∞ 
    —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–µ –±–µ–∑ —Ç–∏–ø–∞ –Ω–∞—Å–µ–ª–µ–Ω–Ω–æ–≥–æ –ø—É–Ω–∫—Ç–∞"""
    for item in point:
        if item in locality_name:
            return item
        
df["locality_name"] = df["locality_name"].apply(fix_locality_name)
df


```
