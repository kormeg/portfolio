Мы видим восходящий тренд. Очевидно, что интервалы сезонности можно будет разглядеть на меньшем отрезке.  
Для задачи анализа ряда на меньшем отрезке создадим функцию

````python
def cut_and_show(first_date, last_date, roll_1, roll_2=None, duration="Some Interval"):
    """Функция принимает на вход первую и последнюю дату интересующего нас отрезка и значение
    размерности окон двух скользящих средних. И выводит их на экран, так же раскладывает на
    тренд и сезонность. Аргумент duration задает название графика для удобочитаемости"""
    period = df[first_date : last_date]
    plt.figure(figsize=(12,6), dpi=100)
    plt.title(duration)
    plt.plot(period)
    plt.plot(period.rolling(roll_1).mean(), c="k", label="rolling mean " + str(roll_1))
    if roll_2:
        plt.plot(period.rolling(roll_2).mean(), label="rolling mean " + str(roll_2))
    plt.legend()
    decomposed_period = seasonal_decompose(period)
    plt.figure(figsize=(12,5), dpi=100)
    plt.title("decomposed " + duration)
    plt.plot(decomposed_period.seasonal, c="g", label="seasonal")
    plt.plot(decomposed_period.trend, c="k", label="trend")
    plt.legend()
    plt.show()
````

<font color='green'>Стационарным нужно делать ряд который предсказываешь, то есть таргеты. Но вместе с этим нужно предусмотреть механизм обратного преобразования предсказаний, что бы на выходе получать количество заказов, а не разницу с пердыдущим часом.</font>

Напишем универсальную функцию для создания Признаков. 

````python
def lag_maker(lag_num, df=df, subtraction=False):
    """Функция добавляет в таблицу отстающие значения временного ряда.
    Если ряд не стационарный, параметр subtraction преобразует отстающие 
    значения в разницы соседних значений"""

    if subtraction:
        for i in range(2, lag_num+1):
            df["lag_{}".format(i)] = df["num_orders"].shift(i-1) - df["num_orders"].shift(i)
    else:
        for i in range(1, lag_num+1):
            df["lag_{}".format(i)] = df["num_orders"].shift(i)
    return df       
````

<font color='blue'><b>Комментарий ревьюера: </b></font> ✔️\
<font color='green'>
1) Признак lag: ДА, это параметр авторегрессии - учёт влияния вчерашних событий на сегодняшние.<br>
2) Признак rolling: ДА, это аналог тренда. Чем больше окно сглаживания, тем ровнее линия тренда <br></font>

<font color='blue'><b>Комментарий ревьюера: </b></font> ❌\

<font color='blue'><b>Комментарий ревьюера 2: </b></font> ✔️\
<font color='green'>Да, lag_1 это значение текущего часа, если мы предсказываем следующий. По условию задачи оно доступно.</font>

````python
def rolling_maker(roll, std=False, df=df):
    """Функция создает скользящие средние и сользящее
    стандартное отклонениедля отстающих значений"""
    df["roll_mean_{}".format(roll)] = df["lag_{}".format(roll)].rolling(roll).mean().shift()
    if std:
        df["roll_std_{}".format(roll)] = df["lag_{}".format(roll)].rolling(roll).std().shift()
    return df
````

````python
def feature_maker(df=df, lag_num=None, subtraction=False, roll_1=None, roll_2=None, roll_3=None, std=False):
    """Функция последовательно объединяет lag_maker и rolling_maker и
    позволяет варьировать конечный результат исходя из задачи"""
    if lag_num:
        if subtraction:
            df = lag_maker(lag_num, subtraction=True)
        else:
            df = lag_maker(lag_num)
            
    if std:
        if roll_1:
            df = rolling_maker(roll_1, std=True)
        if roll_2:
            df = rolling_maker(roll_2, std=True)
        if roll_3:
            df = rolling_maker(roll_3, std=True)
    else:
        if roll_1:
            df = rolling_maker(roll_1)
        if roll_2:
            df = rolling_maker(roll_2)
        if roll_3:
            df = rolling_maker(roll_3)
    return df
````

Полагаю оптимальным количеством отстающих значений будет 24, так же добавим 6-ти, 12-ти, и 24-х часовые скользящие средние и стандартные отклонения. Учитывая нестационарность исходного ряда передадим параметру subtraction значение True

````python
# создаем список для результатов
final_list = []
````

````python
def choose_and_optimize(model_name, model, params, features, target):
    """Функция принимает на вход имя модели, модель, гиперпараметры и обучающие 
    данные проводит кроссвалидацию с подбором параметров и возвращает 
    параметры моделей с наименьшим значением функции потерь"""
    result_list = []
    tscv = TimeSeriesSplit(n_splits=3)
    grid_search = GridSearchCV(model, params, cv=tscv, scoring="neg_root_mean_squared_error", n_jobs=-1)
    model = grid_search.fit(features, target)
    best_params = str(model.best_params_)
    best_score = abs(model.best_score_)
    best_model = model.best_estimator_
    result_list.append(model_name)
    result_list.append(best_params)
    result_list.append(best_score)
    final_list.append(result_list)
    print(result_list) 
    return best_model
````

Нужно передать специальный Сплитер - [TimeSeriesSplit](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.TimeSeriesSplit.html).</s></font>

## Тестирование

Проверим качество лучшей модели с оптимизированными параметрами на тренировочной и тестовой выборках

````python
# список для результатов
testing_list = []
````

````python
def testing(model, X_train=X_train, X_test=X_test, y_train=y_train, y_test=y_test, test=False):
    """Функция принимает на вход необученную модель, тренировочные и тестовые данные,
    производит обучение на тренировочных данных и предсказание по тренировочной или 
    тестовой выборке(параметр test), вычисляет и выводит на экран RMSE"""
    result_list = []
    print("обучение")
    %time model.fit(X_train, y_train)
    print("предсказание")
    if test:
        %time model.predict(X_test)
        predictions = model.predict(X_test)
        rmse = mean_squared_error(y_test, predictions, squared=False)
        aim = "test"
    else:
        %time model.predict(X_train)
        predictions = model.predict(X_train)
        rmse = mean_squared_error(y_train, predictions, squared=False)
        aim = "train"
    result_list.append(aim)    
    result_list.append(str(model))
    result_list.append(rmse)
    testing_list.append(result_list)
    print(f"RMSE: {rmse}")
````
