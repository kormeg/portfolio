Эффективность обогащения рассчитывается по формуле:  


$$ Recovery = \frac{C * (F - T)}{(F * (C - T)} * 100 \% $$

В переводе в строчную запись:  

$$Recovery = ((C * (F - T)) / (F * (C - T))) * 100%$$  
  
  
где:  
  
  
C — доля золота в концентрате после флотации/очистки;  
F — доля золота в сырье/концентрате до флотации/очистки;  
T — доля золота в отвальных хвостах после флотации/очистки.  
___  
  
Для прогноза коэффициента нужно найти долю золота в концентратах и хвостах. Причём важен не только финальный продукт, но и черновой концентрат.  

Метрика качества:  

$sMAPE$ (англ. $Symmetric Mean Absolute Percentage Error$, «симметричное среднее абсолютное процентное отклонение»).
Она похожа на $MAE$, но выражается не в абсолютных величинах, а в относительных. Почему симметричная? Она одинаково учитывает масштаб и целевого признака, и предсказания.
Метрика $sMAPE$ вычисляется так:  

$$ sMAPE = \frac{1}{N}  \sum_{i=0}^{N}\frac{|y_i - \hat y_i|}{(|y_i + \hat y_i|) / 2} * 100\%$$

```python
# for column in df_full:
#     display(df_full[df_full[column] == 0])
    
def zero_destroyer(df):
    """Функция удаляет все строки в которых встречаются нулевые значения
    и показывает насколько при этом уменьшается таблица"""
    len_before = len(df)
    print(len_before)
    for column in df.columns:
        df = df[df[column] != 0] #.reset_index(drop=True)
        display(len(df))
    len_after = len(df)
    print(f"Доля удаленных строк: {1 - len_after/len_before:.2%} процентов")
    return df
```

## Модель

исходя из формулы:  

$$ sMAPE = \frac{1}{N}  \sum_{i=0}^{N}\frac{|y_i - \hat y_i|}{(|y_i| + |\hat y_i|) / 2} * 100\%$$

напишем функцию расчета sMAPE

```python

def smape(y, y_hat):
    """функция принимает на вход значения целевого признака 
    и предсказаний модели и возвращает значение sMAPE"""
    numerator = np.abs(y - y_hat)
    denominator = (np.abs(y) + np.abs(y_hat)) / 2
    
    smape_score = (1 / len(y)) * (numerator / denominator).sum() * 100
    return smape_score
```

```python
# создаем скорер
smape_scorer = make_scorer(smape, greater_is_better=False)
```

для подсчета финального sMAPE нам нужны значения этой величины после первого и последнего этапов. Чтобы их найти построим 2 модели. Исходя из предположения, что очистка на разных этапах может происходить по-разному, для обучения модели первичной очистки возьмем только параметры изначального сырья. для финальной возьмем все признаки

```python
# отделяем целевые признаки в выборках
X_train = df_train.drop(["rougher_output_recovery","final_output_recovery"], axis=1)
y_rougher_train = df_train["rougher_output_recovery"]
y_final_train = df_train["final_output_recovery"]
X_test = df_test.drop(["rougher_output_recovery","final_output_recovery"], axis=1)
y_rougher_test = df_test["rougher_output_recovery"]
y_final_test = df_test["final_output_recovery"]
rougher_columns = [x for x in X_train.columns if "rougher" in x]
X_rougher_train = X_train[rougher_columns]
X_rougher_test = X_test[rougher_columns]
```

```python
# создаем список для результатов
final_list = []

def choose_and_optimize(model_name, model, params, X1, y1, X2, y2):
    """Функция принимает на вход имя модели, модель, гиперпараметры и обучающие 
    данные для 2 этапов проводит кроссвалидацию с подбором параметров и возвращает 
    параметры моделей с наименьшим значением функции потерь"""
    result_list = []
#     model = model(random_state=12345)
    grid_search = GridSearchCV(model, params, cv=5, scoring = smape_scorer )
    best_model = grid_search.fit(X1, y1,)
    best_params = str(best_model.best_estimator_)
    best_rougher_score = abs(best_model.best_score_)
#     model_name = str(model)
    result_list.append(model_name)
    result_list.append(best_params)
    result_list.append(best_rougher_score)
#     result_list.append(model_name)
    
#     model = model(random_state=12345)
    grid_search = GridSearchCV(model, params, cv=5, scoring = smape_scorer )
    best_model = grid_search.fit(X2, y2,)
    best_params = str(best_model.best_estimator_)
    best_final_score = abs(best_model.best_score_)
    result_list.append(best_params)
    result_list.append(best_final_score)
    
    result_smape = 0.25 * best_rougher_score + 0.75 * best_final_score
    result_list.append(result_smape)
    final_list.append(result_list)
    
    return result_list


    
```

Были исследованы разные интервалы параметров моделей и выставлены промежутки в которые попадают лучшие результаты, чтобы модели обучались не слишком долго. Не благодари меня, уважаемый Ревьюер.
